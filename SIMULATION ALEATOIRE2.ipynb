## Exercice 3 : Oscillateur

```python
#3

def eps(k):
    return(np.random.normal(0,k,2))


def X(k,h,w,sigma):
    A=np.array([0,1])
    B=np.zeros((k,2))
    exp_ha=np.array([[np.cos(w*h),(1/w)*np.sin(w*h)],[-np.sin(w*h),np.cos(w*h)]])
    for j in range(k):
        B[j,:]=A
        A=exp_ha@A+eps(sigma)
    return(B)

print(X(10,0.01,1,1))
plt.plot(X(3000,0.01,1,1)[:,0])
plt.plot(X(3000,0.01,1,2)[:,0])

   
    
                                                                  
```

```python
def Y(k,delta):
    B=[]
    A=X(k,0.01,1,1)
    for j in range(k):
        B.append(A[j,0]+np.random.normal(0,delta))
    return(B)

plt.plot(Y(3000,1))
    
        
```
